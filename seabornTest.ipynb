```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('E:/developer/Pokemon.csv', index_col = 0)
df.head()
```

```python
sns.lmplot(x = 'Attack', y = 'Defense', data = df,
          fit_reg = False,
          hue = 'Stage')
```

```python
sns.lmplot(x = 'Attack', y = 'Defense', data = df,
          fit_reg = False,
          hue = 'Stage')
plt.ylim(0,None)
plt.xlim(0,None)
```

```python
sns.boxplot(data = df)
```

```python
status = df.drop(['Total','Stage','Legendary'], axis = 1)
sns.boxplot(data = status)
plt.ylim(0,200)
```

```python
sns.set_style('whitegrid')
sns.violinplot(x = 'Type 1', y = 'Attack', data = status)
```

```python
plt.figure(figsize = (10,6))
sns.swarmplot(x = 'Type 1', y = 'Attack', data = status)
```

```python
plt.figure(figsize = (10,6))

sns.violinplot(x = 'Type 1', y = 'Attack', inner = None, 
               data = df)

sns.swarmplot(x = 'Type 1', y = 'Attack', color = 'k', alpha = 0.3,
              data = df)
plt.title('Attack by Type 1')
```

```python
melted_df = pd.melt(status,
                   id_vars =['Name','Type 1','Type 2'],
                   var_name = 'stat')
```

```python
melted_df.head(6)
```

```python
plt.figure(figsize = (15,10))
sns.swarmplot(x = 'stat', y = 'value', data = melted_df, hue = 'Type 1',split = True)
plt.ylim(0, 260)
plt.legend(bbox_to_anchor = (1,1), loc = 2)
```

```python
corr = status.corr()
sns.heatmap(corr)
```

```python
sns.distplot(df.Attack)
```

```python
sns.countplot(x = 'Type 1', data = status)
plt.xticks(rotation = -45)
```

```python
g = sns.factorplot(x = 'Type 1', y = 'Attack', data = df, col = 'Stage', hue = 'Stage', kind = 'swarm')
g.set_xticklabels(rotation = -45)
```

```python
sns.kdeplot(df.Attack, df.Defense)
```
